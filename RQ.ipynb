```python
import os
import pandas as pd
import smtplib
from email.message import EmailMessage
import numpy as np
from datetime import date
import os
import zipfile
```

```python

def get_excel_file(file):
    
    # list the name of files in the current directory
    file_list = os.listdir()
    print(file_list)

    # get name of the xlsx file
   # xlsx_file_list = [file for file in file_list if '.xlsx' in file]
    xlsx_file_list = file

    if len(xlsx_file_list) < 1: 
        raise ValueError('There is no excel file in the current directory!')
    else:
        xlsx_filename = xlsx_file_list
       
    
     # read in the excel
    df = pd.read_excel(f'./{xlsx_filename}', header=[0],skiprows=3)
    return df
```

```python
def zip_file(folder_path, output_path):
  with zipfile.ZipFile(output_path,'w') as zip_file:
    for filename in os.listdir(folder_path):
      if filename.endswith('.csv'):
        file_path = os.path.join(folder_path,filename)
        zip_file.write(file_path,filename)
```
